````python
print('Mean age of men & women')
#a = df.groupby('Sex').Age.mean()
#display(a)

#using the agg() method which helps to specify multiple aggregation functions at once
df.groupby('Sex').Age.agg(
    Mean_Age=('mean'),
    Std=('std'),
)
````
